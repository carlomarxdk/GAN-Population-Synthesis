```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, Normalizer
```

#### Read the data produced by the 0_Preprocessing Notebook

```python
df = pd.read_pickle('TU_imputed')
df = df.astype(np.int32)
categorical = ['HomeAdrNUTS', 'HomeAdrMunCode',  'RespSex', 'RespPrimOcc', 'RespEdulevel', 'RespHasBicycle',
              'RespHasSeasonticket', 'ResphasDrivlic', 'Handicap', 'HousehAccomodation', 'PosInFamily', 'NuclFamType',
              'HousehAccOwnorRent']
numerical = ['RespYearBorn', 'RespAgeCorrect', 'RespDrivlicYear', 'HousehNumcars', 
            'NuclFamNumPers', 'NuclFamNumAdults', 'NuclFamNumPers1084',
             'NuclFamNumPersO6', 'NuclFamNumDrivLic', 'FamNumPers', 'FamNumAdults',
             'FamNumPers1084', 'FamNumPersO6', 'FamNumDrivLic', 'HousehNumPers',
             'HousehNumAdults', 'HousehNumPers1084', 'HousehNumPersO6',
             'HousehNumDrivLic']
df.head(10)
```

```python
## Convert to categorical Data and see the statistics
df[categorical] = df[categorical].astype('category')
df[categorical].describe().append(df[categorical].isnull().sum().rename('isnull'))
```

#### Normalisation of Data

```python
### normalise the numerical data
norm = Normalizer()
df[numerical] = norm.fit_transform(df[numerical], df[numerical])
df[numerical].head(10)
```

```python
#df.head(10)
```

```python
## Make a copy for further manipulations
df_complete = df 
```

#### Looking at the Correlation

```python
c = df_complete.corr().abs()
s = c.unstack()
so = s.sort_values(kind="quicksort", ascending = False)
so.where(so>0.95 ).dropna()
```

```python
### drop columns based on the correlation metrics
cols_to_drop = ['RespYearBorn', 'FamNumDrivLic', 'NuclFamNumDrivLic',  'FamNumPers', 'NuclFamNumPers', 
               'NuclFamNumPersO6', 'FamNumPersO6', 'NuclFamNumPers1084', 'FamNumPers1084', 'FamNumAdults', 'NuclFamNumAdults']
df_drop = df_complete.drop(cols_to_drop, axis=1)
```

```python
c = df_drop.corr().abs()
s = c.unstack()
so = s.sort_values(kind="quicksort", ascending = False)
so.where(so>0.95 ).dropna()
```

```python
##### Final is a misleading name. This file contains the processed dataset, but the binning is still not pervformed
df_drop.to_pickle('TU_final')
```
